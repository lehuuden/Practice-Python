```python
import random
row=int(input('Enter Row: '))# type 3
column=int(input('Enter Column: '))# type 3
x=[[random.randint(0,0) for i in range(column)] for i in range(row)]
y=x[0]+x[1]+x[2]
def Game_Board(row,column,x):
    n=0
    m=0
    while True:
        a=''
        for i in range(column):
            a=a+' ---'
        print(a)
        n=n+1
        if n==row+1:
            break
        b=''
        w=' '
        for i in x[m]:
            if i==0:
                w=' '
            else:
                w=i
            b=b+'| '+w+' '
        print(b+'|')
        m=m+1
    return
def check(x):
    st=''
    x0=x[0]
    x1=x[1]
    x2=x[2]
    x00=[x[0][0],x[1][0],x[2][0]]
    x11=[x[0][1],x[1][1],x[2][1]]
    x22=[x[0][2],x[1][2],x[2][2]]
    x33=[x[0][0],x[1][1],x[2][2]]
    x44=[x[0][2],x[1][1],x[2][0]]
    z=[x0,x1,x2,x00,x11,x22,x33,x44]
    for i in z:
        if set(i)=={'X'}:
            st='User1 Win'
        elif set(i)=={'O'}:
            st='User2 Win'
        else: continue
    return st
Game_Board(row,column,x)
while 0 in y:
    while 0 in y:
        user1=input('Input your coordinate(user1): ')# coordinate from 1,1 to 3,3 ;example: 2,2
        user1=user1.split(',')
        a1=int(user1[0])-1
        a2=int(user1[1])-1
        if x[a1][a2]==0:
            x[a1][a2]='X'
            y=x[0]+x[1]+x[2]
            Game_Board(row,column,x)
            break
        else:
            print('This position is placed, please again')
            continue
    if check(x)!='':
        print(check(x))
        break
    
    while 0 in y:    
        user2=input('Input your coordinate(user2): ')
        user2=user2.split(',')
        b1=int(user2[0])-1
        b2=int(user2[1])-1
        if x[b1][b2]==0:
            x[b1][b2]='O'
            y=x[0]+x[1]+x[2]
            Game_Board(row,column,x)
            break
        else:
            print('This position is placed, please again')
            continue
    if check(x)!='':
        print(check(x))
        break
if 0 in y:
    print('Congratulation!')
else:
    print('No winer')
```
